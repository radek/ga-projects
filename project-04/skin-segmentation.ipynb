#### Data dictionary

Variable | Column name | Type of Variable | Range | Description | Data type |
---| ---| ---| 
red | R | continuous | 0 - 255 | amount of red shade in skin (additive color model) | int64 |
green | G | continuous | 0 - 255 | amount of green shade in skin (additive color model)  | int64 |
blue | B | continuous | 0 - 255 | amount of blue shade in skin based (additive color model)  | int64 |
material | material | categorical | 1<br />2 | sample of skin<br />sample of non-skin material | int64 |


#### Baseline

```python
skins.material.value_counts(normalize=True)
```

Available color samples describe 21% as skin and remaining 79% as other (non-skin) material.   

#### Improving data set

```python
skins['skin'] = np.where(skins['material'] == 1, 1, 0)
skins.drop('material', axis=1, inplace=True)
```

Creating color in (24 bit format) by calculating `(R * 65536) + (G * 255) + B`

```python
RED_RANGE=65536
GREEN_RANGE=255
BLUE_RANGE=1
skins['color'] = (skins['R'] * RED_RANGE) + (skins['G'] * GREEN_RANGE) + (skins['B'] * BLUE_RANGE)
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(skins['R'], skins['G'], skins['B'], c=skins['skin'], label = 'RGB', alpha=0.2)
ax.set_title("Skin detection in a range of RGB color")
```

#### Setting dimensions

```python
X = skins[['R', 'G', 'B']]
y = skins['skin']
```

#### Scalling

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
X_ss = ss.fit_transform(X)
```

#### Test split and validating logistic regresion model

```python
from sklearn.linear_model import LogisticRegression, Lasso, Ridge, ElasticNet
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import mean_squared_error, classification_report, precision_score, recall_score, confusion_matrix
from helpers import print_predict_scores
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_ss, y)
```

```python
logReg = LogisticRegression()
logReg.fit(X_train, y_train)
```

```python
prediction = logReg.predict(X_train)
```

```python
print_predict_scores(y_train, prediction)
```

Precision remains at level very close to baseline. Reacall performs slightly better comparing to baseline.

```python
print('Train score r2:', logReg.score(X_train,y_train))
print('Test score r2:', logReg.score(X_test,y_test))
```

```python
cross_val_score(logReg, X_train, y_train, cv = 5)
```

```python
cross_val_score(logReg, X_train, y_train, cv = 5, scoring = 'roc_auc')
```

```python
cross_val_score(logReg, X_train, y_train, cv = 5, scoring = 'recall')
```

#### Cross Validation with Lasso

```python
params = [
    {'alpha': [0.1, 0.2, 0.3, 0.5]}
]
```

```python
lasso = Lasso()
grid_search_lasso = GridSearchCV(lasso, params, cv=5, scoring='neg_mean_squared_error')
grid_search_lasso.fit(X_train, y_train)
```

```python
grid_search_lasso.best_estimator_
lasso_model = grid_search_lasso.best_estimator_
lasso_model.fit(X_train, y_train)
```

```python
prediction2 = lasso_model.predict(X_train)
```

```python
mse2 = mean_squared_error(prediction2, y_train)
print('square root :\n{}\n'.format(np.sqrt(mse2)))
```

```python
print('Train score r2:', lasso_model.score(X_train,y_train))
print('Test score r2:', lasso_model.score(X_test,y_test))
```

#### Cross Validation with Ridge

```python
ridge = Ridge()
grid_seaerch_ridge = GridSearchCV(ridge, params, cv = 3, scoring='neg_mean_squared_error')
grid_seaerch_ridge.fit(X_train, y_train)
```

```python
grid_seaerch_ridge.best_score_
ridge_model = grid_seaerch_ridge.best_estimator_
ridge_model.fit(X_train, y_train)
```

```python
prediction3 = ridge_model.predict(X_train)
mse3 = mean_squared_error(prediction3, y_train)
print('square root:\n{}\n'.format(np.sqrt(mse3)))
```

```python
print('Train score r2:', ridge_model.score(X_train,y_train))
print('Test score r2:', ridge_model.score(X_test,y_test))
```

#### Cross Validation against ElastiNet

```python
en = ElasticNet()
grid_seaerch_en = GridSearchCV(en, params, cv = 3, scoring='neg_mean_squared_error')
grid_seaerch_en.fit(X_train, y_train)
```

```python
grid_seaerch_en.best_score_
en_model = grid_seaerch_en.best_estimator_
en_model.fit(X_train, y_train)
```

```python
prediction4 = en_model.predict(X_train)
mse4 = mean_squared_error(prediction4, y_train)
print('square root:\n{}\n'.format(np.sqrt(mse4)))
```

```python
print('Train score r2:', en_model.score(X_train,y_train))
print('Test score r2:', en_model.score(X_test,y_test))
```

#### Interpretation so far  

The best performing model is logistic regression with `0.2852` root square error, `0.7935` precision and `0.8227` recall score. Both precision and recall score are slightly above baseline which indicates fair quality.

#### Precision and Recall curves

```python
from sklearn.metrics import precision_recall_curve, roc_curve, auc
```

```python
y_scores = logReg.decision_function(X_train)
```

```python
precision, recall, thresholds = precision_recall_curve(y_train, y_scores)
```

```python
plt.plot(recall, precision)
plt.title("Recall vs. Precision")
```

```python
params = [
    {'penalty': ['l1', 'l2']}
]
```

```python
for i, metric in enumerate(['precision', 'recall', 'f1', 'roc_auc']):
    grid_cl_loop = GridSearchCV(logReg, param_grid = params, scoring=metric)
    grid_cl_loop.fit(X_train, y_train)
    print('Best score: ', grid_cl_loop.best_score_)
```

```python
fpr_lr, tpr_lr, _ = roc_curve(y_train, y_scores)
roc_auc_lr = auc(fpr_lr, tpr_lr)

plt.figure(figsize = (10, 8))
plt.xlim([-0.01, 1.00])
plt.ylim([-0.01, 1.01])
plt.plot(fpr_lr, tpr_lr, lw=3, label='LogRegr ROC curve (area = {:0.2f})'.format(roc_auc_lr))
plt.xlabel('False Positive Rate', fontsize=16)
plt.ylabel('True Positive Rate', fontsize=16)
plt.title('ROC curve (Cancer Data)', fontsize=16)
plt.legend(loc='lower right', fontsize=13)
plt.plot([0, 1], [0, 1], color='navy', lw=3, linestyle='--')
```
